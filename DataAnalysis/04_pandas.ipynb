```python
# 시리즈 생성
# 1. 리스트로 생성
data = [10,20,30,40]
s = pd.Series(data)
print(s)
```

```python
# 시리즈 주요 속성
data = [10,20,30,40]
s = pd.Series(data, index=['a','b','c','d'], name ='test data')
#name = 시리즈 이름 설정 값

print("values", s.values) # 실제 데이터 배열
print("index", s.index) # 인덱스 객체
print("dtype", s.dtype) # 실제 데이터의 타입
print("size", s.size) #데이터의 개수
print("shape", s.shape) #배열의 형태
print("name",s.name) #시리즈의 이름
```

```python
# 시리즈의 연산
s1 = pd.Series([10,20,30], index=['a','b','c'])
s2 = pd.Series([5,6,7], index=['b','c','d'])

result = s1 + s2
print(result)
```

```python
# 브로드캐스팅
s1 = pd.Series([10,20,30], index=['a','b','c'])

print(s1 + 3)
print(s1 * 3)
```

```python
# 불리언 연산 및 인덱싱
s = pd.Series([50,60,50,80,90,100])

mask = s > 70
print(mask)
print(s[mask])
```

### iLoc, loc

```python
#iloc
# integer Location
# 정수 위치 기반의 인덱싱, 슬라이싱
# numpy의 슬라이싱과 거의 유사함

#단위 행/열 선택
df.iloc[0] #인덱스 지정 [], 첫 번째 행
df.iloc[:,1] #두번째 열'나이'칼럼만 나옴
```

```python
# 여러 행/열 동시 선택
df.iloc[1:4]
df.iloc[:,0:2] #열 첫번째부터 2번째까지 정보
df.iloc[1:4, 0:2]
```

```python
#Fancy Indexing

df.iloc[[0,2,4]] #특정 행 여러개 지정
df.iloc[:,[1,2]] #전체 행에서 1,2열 정보만 갖고와라.[a,b]<열 인덱스 지정
df.iloc[[0,2,4],[1,2]] #특정 행 인덱스+ 특정 열 인덱스
```

```python
# 음수 인덱스 살리이싱
df.iloc[-1]
df.iloc[:,-2:] #끝에서 뒤의 열만 (나이, 직업 나옴)
```

```python
#loc
# location의 약자
# 라벨의 이름 기준으로 인덱싱/ 슬라이싱
# 단일 행/열 선택
#시작과 끝을 모두 포함
# 음수 인덱스 사용 X

df.loc[0]
df.loc[:,"이름"] #모든 행의 이름 칼럼(열) 전부 출력
```

```python
# 여러 행/열 선택
df.loc[2:4,["이름","직업"]]
df.loc[2:4,"이름":"직업"]
```

```python
# 조건식
mask = df["나이"] > 30
df.loc[mask, ["이름","나이"]] #나이 칼럼 기준 30넘는 자만. 이름, 나이 칼럼만 출력
```

### 결측값 처리
- 결측값: 값이 기록되지 않은 상태의 데이터
- NaN: Not a Number

```python
data = {
    "이름": ["서준", "하은", "민준", "서연", "이안", "지민"],
    "나이": [22, 28, np.nan, 31, 27, 24],
    "점수": [89, np.nan, 83, 90, 88, 93]
}
df = pd.DataFrame(data)
df

#이거 데이터 처리할때 numpy 모듈 갖고 와야 한다.
```

```python
# 결측값 탐지
# isnull : 결측값이 맞으면 True 아니면 False
df.isnull() #NaN이 있는 값만 T로 뜸.

# 각 칼럼의 결측값의 수 계산
df.isnull().sum() #칼럼마다 전부 합쳐서 NaN이 몇 개 나오냐 묻는 형태다.

#데이터의 전체 결측값 수 계산
df.isnull().sum().sum() #칼럼까지 전부 합쳤다....같다.

```

```python
# notnull 결측값이 아니면 True, 맞으면 False
df.notnull()
```

```python
# dropna 결측값이 있는 행 삭제
df2 = df.dropna()
df2
```

```python
# 결측값이 있는 열 삭제
df3 = df.dropna(axis= 1) #기본이 axis = 0 (행)
df3
```

```python
# fillna 결측값을 특정 값으로 대체
df4 = df.fillna(0) #NaN을 0으로 대체
df4
```

```python
avg_age = df["나이"].mean()
avg_age

df["나이"] = df["나이"].fillna(avg_age) #df 결측값을, 평균 함수로 구한 값으로 대체하겠다.
df
```

```python
# 이전 값으로 결측값 채우기(forward fill)
df5 = df.ffill()
df5 #0인덱스 점수 89를 갖고와서 채움.
```

```python
#뒤의 값으로 결측값 채우기(backward fill)
df6 = df.bfill()
df6 #2 인덱스 점수 83으로 채움.
```

```python
# 실습
data = {
"도시": ["서울", "부산", "광주", "대구", np.nan, "춘천"],
"미세먼지": [45, 51, np.nan, 38, 49, np.nan],
"초미세먼지": [20, np.nan, 17, 18, 22, 19],
"강수량": [0.0, 2.5, 1.0, np.nan, 3.1, 0.0]
}
df = pd.DataFrame(data)
df
```

```python
#1. ‘미세먼지’컬럼의 평균과 중앙값을 구하세요.
print("평균", df["미세먼지"].mean())


#중앙값
df["미세먼지"].median()
```

```python
#2. ‘초미세먼지’컬럼의 최댓값과 최솟값을 구하세요.

#최대값
print(f"최대값 {df["초미세먼지"].max()}")

#최소값
print(f"최소값 {df["초미세먼지"].min()}")
```

```python
#3. 각 컬럼별 결측값 개수를 구하세요.
df3 = df.isnull().sum()
df3
```

```python
#4. 결측값이 하나라도 있는 행을 모두 삭제한 뒤, 남은 데이터의 ‘초미세먼지’평균을 구하세요.
df4 = df.dropna()
df4
print(df4["초미세먼지"].mean())
#왜 값이 달라졌나요.. df 원본 값 달라져서임...
```

```python
#5. 결측값을 모두 0으로 채운 뒤,‘미세먼지’와 ‘초미세먼지’의 합계를 각각 구하세요.
df5 = df.fillna(0)
df5
print(f"미세먼지 합계 {df5["미세먼지"].sum()}")
print(f"초미세먼지 합계 {df5["초미세먼지"].sum()}")
```

```python
#6. ‘미세먼지’컬럼의 결측값을 평균값으로 채운 뒤, 그 표준편차를 구하세요
m_avg = df["미세먼지"].mean()
print("미세먼지 결측값 평균", m_avg)
df["미세먼지"] = df["미세먼지"].fillna(m_avg)

print("미세먼지 표준 편차", df["미세먼지"].std())
```
