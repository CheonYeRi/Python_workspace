## seaborn
- matplotlib 기반 시각화 라이브러리
- 통계 데이터 시각화에 특화 (pandas와 연계가 좋음)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.__version__
```

```python
# 사용 가능한 데이터셋 이름 출력
sns.get_dataset_names()
```

```python
#데이터셋 로드
#df = sns.load_dataset("tips")
df = sns.load_dataset("titanic")
df.head(10)
```

```python
import matplotlib.font_manager as fm

for f in fm.fontManager.ttflist:
    print(f.name)
```

```python
# 스타일 설정

sns.set_style("ticks")
# 스타일 테마 종류: “white“, “dark“, “whitegrid“, “darkgrid“, “ticks”
sns.set_context("notebook") #문맥에 따라 스타일 크기 조정(현재 기본값)
sns.set_palette("pastel") #컬러 팔레트 테마

plt.rcParams['font.family'] = 'Malgun Gothic' #맑은 고딕이 설치되어 있을 경우 폰트 적용
plt.rcParams['axes.unicode_minus'] = False #마이너스 부호(-) 깨짐 방지
```

```python
# 범주형 데이터 그래프
# countplot

palette = sns.color_palette("pastel")
tips = sns.load_dataset("tips")
sns.countplot(data = tips, x="day", hue="sex", palette=palette)
#data로 갖고올 때 df = sns.load_dataset("tips")에 쓴 걸 기준으로 했다.
plt.title("요일별 방문자수")
plt.grid(linestyle = ":")


plt.show()
#시각 자료형 만들 때 필요한 코드는 사이트 들어가서 검색하면 옵션 값 알려준다...
```

```python
# barplot

palette = sns.color_palette("colorblind")
sns.barplot(data = tips, x = "day", y= "tip", palette=palette)
plt.title('요일별 평균 팁 금액')

plt.show()
```

```python
#displot()
# 히스토그램 + 커널 밀도(kde)
# kde = True 로 하면 kde 그래프 그려준다
sns.displot(data = tips, x= "total_bill", bins =20)
plt.title("Total bill")

plt.show()
```

```python
# scatterplot()
# 산점도

sns.scatterplot(data = tips, x= "total_bill", y= "tip", hue= "sex")

plt.show()
```

```python
#실습
#titanic 데이터 셋 이용한 barplot, displot, scatterplot 활용

# barplot
palette = sns.color_palette("deep")
sns.barplot(data = df, x = "pclass", y= "survived", palette=palette)

plt.show()
```

```python
#displot
sns.displot(data = df, x= "class", kde= True)

plt.show()
```

```python
#scatterplot

sns.scatterplot(data = df , x= "class", y="age", hue="sex")
plt.show()
```

```python
# heatmap()
# 상관계수행렬의 그래프

penguins = sns.load_dataset("penguins")
print(penguins.head(10))

# 상관계수행렬
# 상관계수: 두 변수 사이의 직선 관계가 얼마나 강한지를 숫자로 표현한 값
# heatmap 에서 색으로 표시되는 숫자가 상관계수
# 1: 양의 상관 관계 (하나가 증가하면 다른 것도 같은 비율로 증가)
# ex) 키와 몸무게
# -1: 음의 상관관계 (하나가 증가하면 다른 변수는 감소)
# ex) 공부시간과 실수 개수

corr = penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap= "inferno")
#corr(지정 데이터), fmt(수치 소수점 어디까지), cmap(컬러팔레트), annot =숫자표시여부
plt.show()
```

```python
#jointplot()
sns.jointplot(data= penguins, x= "flipper_length_mm", y= "body_mass_g",kind="scatter")
#kind = 시각 데이터 종류 설정
plt.show()
```

```python
#실습
#penguins 데이터 셋 이용한 heatmap, jointplot 활용 (종별로 분류해보기)

sns.jointplot(data= penguins, x= "flipper_length_mm", 
              y= "body_mass_g", hue="species", kind="scatter")
plt.xlabel("팔 길이")
plt.ylabel("체중")

plt.show()
```
