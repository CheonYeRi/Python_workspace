- 지도 시각화 라이브러리

```python
import folium
import pandas as pd
```

```python
# 기본 지도 생성
m = folium.Map(location=[37.61120705991477,126.91727650972385], zoom_start=18)

# 기본 마커
folium.Marker([37.61120705991477,126.91727650972385],
              popup="Subway",
              tooltip= "구산역",
              icon=folium.Icon(color="black", icon="fa-solid fa-bus",icon_color="red", prefix="fa-solid")
              ).add_to(m)
#popup= 클릭시 뜨는 명칭, tooltip = "이름", icon 이미지는 웹에서 가지고 옴. 

#원형 마커
folium.CircleMarker(
    [37.6103,126.9128],
    radius = 100,
    color = "#4C6EBFFF",
    fill_color = "#91B2FFFF",
    popup = 'CircleMarker'
).add_to(m)

#클릭한 곳에 마커추가
m.add_child(folium.ClickForMarker(popup='내가 클릭한 곳'))

#클릭한 곳의 위도와 경도 표시
m.add_child(folium.LatLngPopup())

m
```

```python
#실습
#강의장 주변 맛집, 카페, 공원등등 원하는 곳 3군데 이상 표시하기

m = folium.Map(location=[37.61025337406213, 126.912908703399], zoom_start=18)

folium.Marker([37.61025337406213, 126.912908703399],
              popup="CodingOn",
              tooltip= "강의실",
              icon=folium.Icon(color="blue",icon='star',icon_color="black")
              ).add_to(m)

folium.Marker([37.609533849475184, 126.91464234743492],
              popup="춘리마라탕",
              tooltip= "음식점",
              icon=folium.Icon(color="blue",icon='star',icon_color="black")
              ).add_to(m)

folium.Marker([37.61170014612301, 126.91698432287136],
              popup="다이소 구산역점",
              tooltip= "다이소",
              icon=folium.Icon(color="blue",icon='star',icon_color="black")
              ).add_to(m)

m.add_child(folium.ClickForMarker(popup='내가 클릭한 곳'))

m.add_child(folium.LatLngPopup())

m
```

```python
#택시승차대 위치 표시하기
df = pd.read_csv("./dataset/택시승차대 현황.csv", encoding='cp949')
#인코딩 오류 떠서 임의로 지정
df.head(10)

taxi = df.loc[:,["위치명","위도","경도"]]

taxi
```

```python
latitude = taxi["위도"].mean()
longitude = taxi["경도"].mean()


m = folium.Map(location=[latitude,longitude], zoom_start=15)

for idx, row in taxi.iterrows():
    #print(idx,row["위치명"],row["위도"],row["경도"])
    folium.Marker(
        [row["위도"],row["경도"]],
        tooltip=row["위치명"],
        icon=folium.Icon(color="orange", icon="star", icon_color="black",prefix="fa-solid")
    ).add_to(m)

m
```
