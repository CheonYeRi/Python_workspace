## Beautiful Soup
- HTML과 XML 파일에서 데이터를 추출하기 위한 라이브러리

```python
from bs4 import BeautifulSoup
import requests
```

```python
#html 작동 원리 !누르면 양식 뜨는데 컨트롤 스페이스 누르면 도로 떠줌. 자동완성 떠줘야 나옴.
#F12 단축기 : 개발자 도구. html 양식 나와줌. 빠르게 변경사항 확인 되도록 해주는 거라 새로고침시 원형으로 돌아감.
#코드 양식 중 바디 영역만 보여준다. 

with open('test.html','r', encoding="utf-8") as f:
    html_data = f.read()

#soup 객체 생성
#soup = BeautifulSoup(html_data,"html.parser") #내장 파서
soup = BeautifulSoup(html_data,"lxml") #xml일 때 사용, 설치 필요
#print(soup)
#print(soup.prettify()) #들여쓰기 표시

#파서 차이 비교
print(BeautifulSoup("<a></p>","html.parser")) #<a></a>

print(BeautifulSoup("<a></p>","lxml")) #<html><body><a></a></body></html>
```

```python
#데이터 선택
#find() - 첫번째 매칭 요소 선택
#1) 태그를 기준으로 탐색
title_tag = soup.find("h1")
print(title_tag)
print(title_tag.text)
print(title_tag.get_text())
```

```python
#find()- 속성 조건으로 검색 가능
result = soup.find("h1", class_="sub_title")
print(result.text)
```

```python
# find_all() - 모든 매칭요소 선택
```

```python
#select() - 모든 매칭 요소 선택
# CSS 선택자로 선택
result = soup.select("ul.items")
result
```

```python
#select_one() = 첫번째 매칭 요소 선택
result = soup.select_one("ul.items")
result
```

### Requests
- HTTP 프로토콜을 이용하여 웹 사이트로부터 데이터를 송수신하는 라이브러리

```python
#beautifulsoup & requests 함께 이용

# 멜론에서 Top10의 노래 제목을 받아오기

url = "https://www.melon.com/chart/index.htm"
headers = {
    "User-Agent":"Mozilla/5.0"
}

response = requests.get(url,headers= headers)
# print(response.status_code) # 200: 성공, 404: 페이지 없음, 500: 서버 에러
# print(response.text)

soup = BeautifulSoup(response.text, "lxml")

songs = soup.select("div.ellipsis.rank01 a")[:10]

for idx, song in enumerate(songs):
    print(f"{idx+1}. {song.text}")
```

```python
#실습 웹 크롤링
#사용자에게 검색어를 입력 받아 검색된 뉴스의 제목과 링크 가져와보세요

url = "https://news.naver.com/"
headers = {
    "User-Agent":"Mozilla/5.0"
}
#query = input("검색 키워드: ")

news = requests.get(url, headers=headers)
soup = BeautifulSoup(news.text, 'html.parser')
print(soup)

news_log = soup.select("div.sds-comps-vertical-layout sds-comps-full-layout")

for log in news_log:
    title = log.get_text().strip()
    link = log.find("div", class_="sds-comps-vertical-layout sds-comps-full-layout DlhZas37Se4xSUxh6MWQ")
    print(f"{title} : {link}")
```
