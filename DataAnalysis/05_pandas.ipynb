## 조건 필터링

```python
data = {
    '이름': ['홍길동','김철수','이영희','박민수'],
    '점수': [90, 70,95, 85],
    '반' : [2,1,1,2]
}

df = pd.DataFrame(data)
df
```

```python
mask = df["점수"] >= 80
filtered = df[mask]
filtered
```

```python
# 여러 조건 결합
# and : &
# or : |
# not : ~

# & 예시
df[(df["반"] == 1)& (df["점수"] >= 80)]
```

```python
# | 예시
df[(df["반"] == 2) | (df["점수"] >= 90)]
```

```python
# ~ 예시
df[~(df["점수"] >= 80)]
```

```python
#isin()
# 여러 값 중에 하나에 해당하는 지 여부를 판단할 때 사용

#이름이 홍길동이거나 박민수 이거나 한 경우
mask = [(df["이름"] == "홍길동") | ((df["이름"]) == "박민수")]
#mask = df["이름"].isin( ["홍길동", "박민수"])
mask
```

### 정렬
- 값 기준 정렬: sort_valuse
- 인덱스 기준 정렬: sort_index

```python
#값 기준 정렬: sort_valuse
data = {
    '이름': ['홍길동','김철수','이영희','박민수'],
    '점수': [90, 70,85, 95],
    '반' : [2,1,1,2]
}

df = pd.DataFrame(data)
df
```

```python
#오름차순 정렬
df2 = df.sort_values("점수").reset_index(drop=True)
df2
```

```python
#내림차순 정렬
#ascending = False
df3 = df.sort_values("점수",ascending=False).reset_index(drop=True)
df3
```

```python
#여러 기준으로 정렬
df4 = df.sort_values(["반","점수"], ascending=[True,False]).reset_index(drop=True)
#반을 기준으로 오름차순, 점수 기준으로 내림차순, 인덱스는 재정렬
df4
```

```python
#원본 변경
df.sort_values(["반","점수"],ascending=[True,False],inplace=True)
#inplace =True :원본 변경
df1 = df.reset_index(drop=True)
df1
```

```python
# 인덱스 기준 정렬: sort_index
df_shuffled = df.sample(frac=1, random_state=42) #정렬 섞기
df_shuffled

#행 인덱스 기준으로 정렬
df5 = df_shuffled.sort_index() #오름차순
df5

#열 이름 기준 정렬(알파벳순)
df6 = df.sort_index(axis=1)
df6
```

### groupby

```python
df = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'B', 'C'],
    'name': ['Kim', 'Lee', 'Park', 'Choi', 'Jung', 'Han'],
    'score': [90, 85, 80, 70, 95, 88]
})
df
```

```python
# 단일 칼럼 기준 그룹화
grouped = df.groupby("team")

#집계함수 적용
result_mean = grouped["score"].mean()
print(result_mean)

# 개수
result_count = grouped["score"].count()
print(result_count)
```

```python
df2 = pd.DataFrame({
    'team': ['A', 'A', 'B', 'B', 'B', 'C'],
    'gender': ['M', 'F', 'F', 'M', 'M', 'F'],
    'score': [90, 85, 80, 70, 95, 88],
    'age' : [21, 22, 23, 25, 20, 27]
})
df2
```

```python
#여러 칼럼 기준 그룹화
result = df2.groupby(["team","gender"])["score"].mean()
result
#팀, 젠더 기준으로 그룹화 하고/ 스코어 평균값을 낸다. 뜻이다.
```

```python
# as_index=False 옵션
# 그룹 라벨이 인덱스로 설정됨.
result = df2.groupby("team",as_index=False)["score"].sum()
result
#인덱스 표기 생기고, 점수 기준으로 합산
```

```python
#여러 집계 함수를 한번에 적용: agg()
result = df2.groupby("team", as_index=False)["score"].agg(["sum","mean","count"])
result
```

```python
#그룹별로 여러 칼럼에 다른 집계 함수 적용
result = df2.groupby("team").agg({
    "score": "mean",
    "age" : "max"
})
result
#agg()에서 딕셔너리 {}처럼 지정 적용하면 된다.
```

```python
df = pd.DataFrame({
'class': [1, 1, 1, 2, 2, 2],
'subject': ['Math', 'Math', 'Eng', 'Math', 'Eng', 'Eng'],
'score': [80, 90, 85, 70, 95, 90]
})

#2. 아래DataFrame에서 반(class)별, 과목(subject)별로 
# 시험에 응시한 학생수(count)와평균점수(avg)를구하세요
df2 = df.groupby(["class","subject"])["score"].agg(["mean","count"])
print(df2)
```
