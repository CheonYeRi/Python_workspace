## 11. selenium
- 웹 자동화를 위한 라이브러리

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

import time
```

```python
# Chrome 브라우저 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 원하는 페이지로 이동
driver.get("https://www.google.com")
time.sleep(3) #딜레이 추가
driver.get("https://www.naver.com")
time.sleep(3)

#페이지 이동
driver.back() #뒤 페이지 이동
time.sleep(3)
driver.forward() # 앞 페이지 이동
time.sleep(3)
driver.refresh() #현재 페이지 새로고침
time.sleep(3)

# 크기 변경
driver.set_window_size(1200,800)
time.sleep(3)
#최대화
driver.maximize_window()
time.sleep(3)
#최소화
driver.minimize_window()
time.sleep(3)
#창 배치
driver.set_window_position(50,50)
time.sleep(3)


driver.quit() #브라우저 종료
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.selenium.dev/selenium/web/web-form.html")

#개발자도구F12창에서 ctrl+F하면 검색창이 뜬다. (클래스, 명칭 검색할 때 씀)

#웹 페이지의 요소 선택과 제어
text_input = driver.find_element(By.ID, "my-text-id")

time.sleep(3)

#요소 제어
text_input.send_keys("Hello Selenium") #요소 입력
time.sleep(3)
text_input.clear() #지우기
time.sleep(3)
text_input.send_keys("Coding")
time.sleep(3)

submit_btn = driver.find_element(By.TAG_NAME,"button")
submit_btn.click()
time.sleep(5)

driver.quit()
```

```python
#구글 검색 예제

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.google.com/?hl=ko&zx=1765512080522&no_sw_cr=1")

time.sleep(3)
search_box = driver.find_element(By.NAME,"q") #By.TAG, "textarea"도 됨
search_box.send_keys("Python", Keys.RETURN)

time.sleep(10)

driver.quit()
```
