```python
# ndarry의 생성 
# 쉬프트 + 엔터: 실행

a = np.array([[1,2,3],[4,5,6]])
b = np.array([1.0,3.14,1,24])

a

# 배열의 구조
print(f"배열의 구조: {a.shape}")

# 배열의 차원 수
print(f"배열의 차원: {a.ndim}")

#데이터 타입
print(f"배열의 데이터 타입: {a.dtype}")
print(f"배열의 데이터 타입: {b.dtype}")

#형변환
new_a = a.astype(np.float64)
print(f"수정한 배열의 데이터 타입: {new_a.dtype}")

```

### 1-5. 인덱싱과 슬라이싱
- 다차원 배열을 다루는 편의 기능 제공
- Python의 시퀸스보다 빠름

```python
#인덱싱 a[]<인덱싱 좌표
a = np.array([10,20,30,40,50])
print(a[2])
print(a[-1])
```

```python
# 다차원 인덱싱
#파이썬 리스트
matrix = [[1,2,4], [4,5,6]]
print("파이썬 인덱싱:",matrix[1][1])

#numpy 배열
a2 = np.array([[1,2,4],[4,5,6]])
print("numpy 인덱싱:", a2[1,1])

#3차원 배열 인덱싱
a3 = np.arange(24).reshape(2,3,4)
print(a3)
print("numpy 인덱싱:", a3[1,1,1])
```

```python
# 슬라이싱
# arr[행 슬라이스, 열 슬라이스]
# arr[..., 4차원 슬라이스, 3차원 슬라이스, 2차원, 1차원]

a1 = np.array([10,20,30,40,50])
print(a1[1:3])
print(a1[2:])
print(a1[:2])
print(a1[:])
print(a1[::2])
print(a1[::-1])
```

```python
# 파이썬 리스트와의 차이
# 파이썬 리스트
py_list = [10,20,30,40,50]
sliced = py_list[1:4]
sliced[1] = 100
print("py 원본",py_list)
print("py 슬라이싱", sliced)
print()

#numpy 배열
a1 = np.array([10,20,30,40,50])
a1_silced = a1[1:4]
a1_silced[1] = 100
print("numpy 원본",a1)
print("numpy 슬라이싱",a1_silced)
```

```python
# 얕은 복사: 복사본이 원본과 메모리를 공유 > 변경 사항이 서로에게 영향을 줌

b1 = np.array([1,2,3])
b1_viewd = b1.view()
b1_viewd[1] = 10
print("원본",b1)
print("복사본", b1_viewd)
print()

# 깊은 복사: 복사본이 원본과 독립적으로 복사됨 > 서로 영향 주지 않는다.

b2 = np.array([1,2,3])
b2_viewd = b2.copy()
b2_viewd[1] = 20
print("원본", b2)
print("복사본", b2_viewd)
print()
```

```python
# Fancy Indexing
# 정수 배열을 사용하여 여러 인덱스로 여러 요소를 한번에 선택
af = np.arange(1,21)
print(af)
print(af[[4,7,1]])

af2 = np.arange(1,21).reshape(4,5)
print(af2)
print(af2[[1,3],[2,4]])
```

```python
# boolean Indexing
ab = np.linspace(10,100,10)
print(ab)
print(ab[ab>40])
print()

#boolean masking
ab2 = np.arange(0,21)
print(ab2)
mask = ab2 % 2 == 0
print(mask)
print(ab2[mask])
```

```python
#실습
#5 4x5 배열에서 가운데 2x3 부분을 슬라이싱한 뒤 copy()를 이용해 독립 배열을 만드세요
arr= np.arange(1, 21).reshape(4, 5)
print(arr)

arr_viewd = arr.copy()
arr_viewd = arr[1:3, 1:4]
print("원본", arr)
print("복사본", arr_viewd)

#모범 답안
arr_viewd = arr[1:3,1:4].copy
```

```python
#6  3x4 배열에서 짝수이면서 10 이상인 값만 선택하세요.(&을활용)

arr= np.array([[ 4,  9, 12,  7], [10, 15, 18,  3], [ 2, 14,  6, 20]])
print(arr[arr>10])

#모범 답안
print(arr[(arr % 2 == 0) & (arr >= 10)])
```

```python
#7 5x5 배열에서 2, 4번째 행을 선택하고, 선택된 행에서 열 순서를 [4, 0, 2]로 재배치

arr= np.arange(1, 26).reshape(5, 5)
print(arr)

arr_1 = arr[[1,3]]
print(arr_1[::-1,[4,0,2]])

#모범 답안
arr_1 = arr[[1,3]]
arr_1[:,[4,0,2]]
```

```python
#8 5x3 배열에서 각 행의 첫 번째 값이 50 이상인 행만 Boolean Indexing으로 선택

arr= np.array([[10, 20, 30], [55, 65, 75], [40, 45, 50], [70, 80, 90], [15, 25, 35]])
arr[arr>50].reshape(2,3)

#모범 답안
arr[arr[: ,0] >= 50 ]
```

```python
# 0~9 난수로 이뤄진 (4,4)배열 형성하고 인덱싱으로 추출(for사용)
# 문제 이해가 잘 안 간다...
# 전부 못 풀어서 생략
a4 = np.random.randint(0,10,(4,4))
print(a4)
print(a4[[0,1,2,3],[0,1,2,3]])
print(a4[[0,1,2,3],[3,2,1,0]])


answer = []
answer2 = []
for i in range(4):
    answer.append(a4[i][i])
    answer2.append(a4[i][3 - i]) #인덱스 뒤로 하는 좌표값 지정
print("주대각선:\n",answer)
print("주대각선:\n",answer2)
```
