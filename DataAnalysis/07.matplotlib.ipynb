### 그래프 그리기

### 그래프 구성요소(1)
- figure: 여러 그래프가 그려지는 최상위 컨테이너
- axes : 실제 그래프가 그려지는 공간
- axis : 그래프의 축

```python
#subplots()
# 여러개의 그래프 그릴 때 사용
# figure, axes 동시 반환
# fig, axs = plt.subplots(행,열)
fig, axs = plt.subplots() #하나의 figure와 하나의 axes 만듬
```

```python
# 1행 2열의 axes 생성
fig, axs = plt.subplots(1,2) #1행 (가로) 2열(세로 2개) 모양이다.

axs[0].plot([1,2,3],[1,2,3]) #x,y 값 설정이지 이건 행, 열과 다르다.
axs[1].plot([1,2,3],[1,4,9])

plt.show()
```

```python
# 2행 2열의 axes 생성
fig, axs = plt.subplots(2, 2)
a = [1,2,3]
b = [1,4,9]

fig.suptitle("여러 함수 그래프")

axs[0,0].plot(a, a)
axs[0,0].set_title("1번 그래프")

axs[0,1].plot(a, b)
axs[0,1].set_title("2번 그래프")

axs[1,0].plot(a, [3,2,1])
axs[1,0].set_title("3번 그래프")

axs[1,1].plot(a, [9,4,1])
axs[1,1].set_title("4번 그래프")

plt.tight_layout() #그래프 간격 자동 조절

plt.show()
```

### 그래프의 구성요소(2)
- 제목, 레이블, 범례, 그리드

```python
# title
plt.plot([1,2,3,4,5], [1,4,9,16, 25], label = '제곱 그래프', linestyle = '--', color = "c", marker = "^")
plt.title("제곱 함수 그래프", color ="b", fontsize = 20)

#xlabel, ylabel
plt.xlabel("x값")
plt.ylabel("y값(x의 제곱)")
# 라벨 : 축의 이름 붙이기

#범례(legend)
plt.legend()

#격자(grid)
plt.grid(linestyle = '--')

plt.show()
```

```python
# 하나의 axes에 여러개의 그래프 그리기
x = np.arange(0,10)
y1 = 2 * x + 1
y2 = 3 * x - 5

plt.plot(x, y1, marker = 'o', label = "y = 2x + 1", color = "c")
plt.plot(x, y2, marker = '*', label = "y = 3x - 5", color = "m")

plt.title("두개의 선 그래프 비교")
plt.xlabel("x값")
plt.ylabel("y값")
plt.legend()
plt.grid(linestyle = ":", alpha = 0.5)

plt.show()
```

```python
#실습
#아래 데이터를 이용해서 예시와 같이 선그래프를 그려보세요

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov',"Dec"]
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

x = np.arange(12)

plt.plot(months, sales_2019, label = "2019년도", color = "b")
plt.plot(months, sales_2020, label = "2020년도", color = "r")

#plt.set_xticks(x,"months")
plt.title("Monthly Sales Comparison(2019-2020)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc ="lower right", title = "판매 연도")
plt.grid(linestyle = ":", alpha = 0.5)

plt.show()
```

```python
# 막대 그래프

labels = ["A","B","C","D"]
values = [10,24,15,32]

#plt.bar(labels,values) #수직
plt.barh(labels,values) #수평

plt.show()
```

```python
# 여러 막대 그래프
labels = ["1분기","2분기","3분기","4분기"]
men_means = [20,34,30,35]
women_means = [25,32,34,40]

x = np.arange(len(labels))

fig, ax = plt.subplots()
width = 0.4

ax.bar(x - width/2, men_means, width, label ="남성")
ax.bar(x + width/2, women_means, width, label = "여성")

ax.legend() #컬러따라 남성,여성 라벨 표시

ax.set_xticks(x,labels) #숫자 대신 문자열을 표시

plt.show()
```

```python
# 실습 2
# 막대 그래프 그리기
#아래 데이터를 바탕으로 예시와 같이 막대그래프를 그려보세요

categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45] 

x = np.arange(len(categories))

fig, ax = plt.subplots()
width = 0.4

ax.bar(categories, data, width)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar chart")
ax.set_xticks(x,categories) #숫자 대신 문자열을 표시 (x는 5개 요소로 측정됨)
plt.grid()
plt.ylim(0,50) #y축을 0~50범위로 조정
plt.xticks(rotation=50) #X라벨의 모양을 50도 회전
plt.yticks(rotation=50)

plt.tight_layout()

plt.show()
```

```python
#히스토그램
#랜덤 데이터 생성

data = np.random.randn(1000) #평균 0 표준편차 1인 정규분포

plt.hist(data,bins=30, color= "r", edgecolor = "black")
#bins : 구간의 개수, 구간의 리스트
#color, edgecolor: 막대의색상, 테두리색상
plt.show()
```

```python
#실습
#히스토그램 그리기
#임의로 1~6 사이의 주사위눈 100개를 생성해서 각 숫자가 얼마나 자주 나왔는지 히스토그램을 그려보세요
data = np.random.randint(1,7,size=100)

cnt, bins, _ = plt.hist(data, bins= np.arange(1,8) - 0.5 , color= "b", rwidth = 0.5, edgecolor = "black")
print(bins) #bins의 간격 설정을 보여주기 위한 출력
# _는 변수 넣는 자리를 쓰지 않는다. 뜻으로 _ 넣음.

plt.xlabel("주사위 눈")
plt.ylabel("빈도")
plt.title("주사위 눈 빈도 히스토그램")
plt.yticks(rotation=90)

plt.show()
```

```python
# 산점도
x = np.linspace(2,30,20)
y = np.linspace(10,40,20)

y_noisy = y + np.random.uniform(-7,7,20)

plt.scatter(x, y_noisy)
plt.grid(linestyle = ":")

plt.show()
```

```python
#실습
#산점도 그리기
#x값은 0부터 10까지 30개의 균등한 간격으로 구성된 수열을 만들고,
# y값은 y = 3x + 2 + 난수(정규분포, 평균0, 표준편차3)로 생성하여 산점도를 그려보세요

x = np.linspace(0,10,30)
y = (3 * x + 2) + np.random.uniform(-3,3,30)
#noise = np.random.normal(0,3,30) #난수: 평균 0 표준편차 3, 개수 30
#y = 3 * x + 2 + noise < y값 모범 답안이다

plt.scatter(x, y, s = 10)
plt.xlabel("x 값")
plt.ylabel("y 값")
plt.title("노이즈가 추가된 산점도")
plt.ylim(0,35)
plt.grid(True, linestyle = ":")

plt.show()
```
